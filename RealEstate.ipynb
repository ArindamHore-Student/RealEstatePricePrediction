```python
# For learning purposes
import numpy as np
def split_train_test(data, test_ratio):
    """
    Splits the given dataset into training and test sets.

    Parameters:
    - data: pandas DataFrame
      The dataset to split into training and test sets.
    - test_ratio: float
      The proportion of the dataset to be used as the test set (e.g., 0.2 for 20%).

    Returns:
    - test_set: pandas DataFrame
      The portion of the data to be used as the test set.
    - train_set: pandas DataFrame
      The portion of the data to be used as the training set.
    """

    # step 0: Set a random seed to ensure reproducibility of results
    np.random.seed(42)
    
    # Step 1: Randomly shuffle the indices of the dataset
    shuffled_indices = np.random.permutation(len(data))

    # Step 2: Calculate the size of the test set based on the given ratio
    test_set_size = int(len(data) * test_ratio)

    # Step 3: Split the shuffled indices into test and training indices
    test_indices = shuffled_indices[:test_set_size]  # First part for the test set
    train_indices = shuffled_indices[test_set_size:]  # Remaining part for the training set

    # Step 4: Use the indices to create the test and training datasets
    test_set = data.iloc[test_indices]
    train_set = data.iloc[train_indices]

    # Step 5: Return the test and training datasets
    return test_set, train_set
```

```python
# ## Feature Scaling
# Feature scaling is a crucial preprocessing technique in machine learning that normalizes the range of independent variables or features in a dataset. It ensures that all features contribute equally to the model's learning process and performance.

# ### Why Feature Scaling is Important
# 1. **Improved Algorithm Performance**: Many machine learning algorithms, especially those based on gradient descent, converge faster and find optimal solutions more easily with scaled features.
# 2. **Unbiased Feature Contribution**: Scaling prevents features with larger magnitudes from dominating the learning process, allowing all features to contribute proportionately.
# 3. **Enhanced Distance-Based Algorithms**: Algorithms like k-Nearest Neighbors (k-NN), K-Means Clustering, and Support Vector Machines (SVMs) rely on distance calculations and benefit significantly from scaled features.
# 4. **Numerical Stability**: Scaling helps prevent numerical instability issues in computations, particularly for algorithms sensitive to the scale of input features.
# 5. **Model Interpretability**: With scaled features, it becomes easier to interpret the relative importance of each feature in the model.

# ### Common Feature Scaling Techniques
# 1. **Normalization (Min-Max Scaling)**:
#    Scales features to a fixed range, typically between 0 and 1.
#    Formula: X_scaled = (X - X_min) / (X_max - X_min)
# 2. **Standardization (Z-score Normalization)**:
#    Transforms features to have a mean of 0 and standard deviation of 1.
#    Formula: X_scaled = (X - mean(X)) / std(X)
# 3. **Robust Scaling**:
#    Uses statistics that are robust to outliers, like median and interquartile range.
# 4. **Log Scaling**:
#    Applies logarithmic transformation to compress a wide range of values.

# ### When to Apply Feature Scaling
# Feature scaling is particularly important for:
# - Gradient descent-based algorithms
# - Distance-based algorithms (e.g., k-NN, K-Means)
# - Support Vector Machines
# - Neural Networks
# - Regularization techniques

# However, tree-based algorithms like Random Forests and Decision Trees are generally less sensitive to feature scaling.
# By applying appropriate feature scaling techniques, machine learning practitioners can significantly improve model performance, convergence speed, and overall predictive accuracy.
```

```python
# Pipelines
# A machine learning pipeline is indeed a series of interconnected steps that automate and streamline the process of building and deploying machine learning models.
# It organizes the flow of data through various stages of the machine learning workflow.
```

```python
# Importing necessary libraries
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Creating a pipeline with two steps:
# 1. 'imputer': Handles missing values by replacing them with the median value of each column.
# 2. 'std_scaler': Standardizes features by removing the mean and scaling to unit variance.

df_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),  # Step 1: Impute missing values
    ('std_scaler', StandardScaler()),               # Step 2: Standardize the data
    # Add more steps as per requirement
])

# Apply the pipeline to the DataFrame (df)
df_transformed = df_pipeline.fit_transform(df)

# Explanation:
# - The pipeline first fits the transformers (like imputer and scaler) to the data in 'df'.
# - Then, it transforms the data using the fitted transformers.
# - 'df_transformed' contains the processed data after all transformations in the pipeline have been applied.
```

```python
## Selecting a Desired Model for my dataset
```

```python
# Importing necessary libraries
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Training a Linear Regression model
# model = LinearRegression() 
# model = DecisionTreeRegressor() 
model = RandomForestRegressor()
model.fit(df_transformed, df_labels)

# Explanation:
# - The Linear Regression model is trained using the transformed features ('df_transformed') and target labels ('df_labels').
# - The 'fit' method fits the model to the data, finding the best-fit line for prediction.
```

```python
# Testing the model
# Select a sample of the data for testing
df_sample = df.iloc[:5]  # First 5 rows as test features
df_sample_labels = df_labels.iloc[:5]  # First 5 rows as test labels

# Transform the sample data using the existing pipeline
prepared_data = df_pipeline.transform(df_sample)

# Explanation:
# - 'df_sample' contains the first 5 rows of features from the dataset.
# - 'df_sample_labels' contains the corresponding target labels.
# - 'prepared_data' contains the transformed test data after applying the pipeline (scaling and imputation).
```

```python
# Importing necessary libraries
from sklearn.metrics import mean_squared_error
import numpy as np

# Analyzing Errors
# Generate predictions for the entire dataset
df_predictions = model.predict(df_transformed)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(df_labels, df_predictions)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Explanation:
# - 'df_predictions' contains the predicted values for the dataset.
# - 'mean_squared_error' calculates the average squared difference between actual and predicted values.
# - 'rmse' is the square root of 'mse', providing an error measure in the same units as the target variable.
```

```python
# Better Evaluation Technique aka Cross Validation
```

```python
# Import necessary libraries
from sklearn.model_selection import cross_val_score
import numpy as np

# Perform cross-validation with the model, transformed data (df_transformed), and labels (df_labels)
# Using 'neg_mean_squared_error' as scoring metric, as scikit-learn returns negative values by default for loss metrics.
model_scores = cross_val_score(model, df_transformed, df_labels, scoring="neg_mean_squared_error", cv=10)
# Since 'cross_val_score' returns negative MSE, we take the negative and compute the square root to get RMSE
rmse_model_scores = np.sqrt(-model_scores)

# Print the RMSE scores across folds
print(rmse_model_scores)
```

```python
# When to Use Each Approach:

# Single Model Evaluation is sufficient when you're just interested in a quick check of your model's performance on a dataset.
# This method is common when you have already done model validation in other ways and just want to see the error on your entire dataset.

# Cross-Validation is preferable when you need a more thorough, reliable performance metric.
# It is especially helpful when tuning models or comparing models, as it reduces the risk of a misleading result from a single, potentially unrepresentative data split.
```

```python
def print_scores(scores):
    """
    This function takes a collection of scores (e.g., model evaluation metrics), and prints the scores, 
    their mean, and their standard deviation.

    Parameters:
    scores (array-like): A collection of numerical values, such as an array or list of model evaluation scores.
    
    The function performs the following:
    1. Prints the individual scores provided in the input.
    2. Calculates and prints the mean of the scores (average score).
    3. Calculates and prints the standard deviation of the scores, providing an indication of the variability of the scores.
    """
    # Print all the scores in the input
    print("Scores: ", scores)
    
    # Print the mean (average) of the scores
    print("Mean: ", scores.mean())
    
    # Print the standard deviation of the scores (measure of dispersion)
    print("Standard Deviation: ", scores.std())

# Example usage of the function:
# Print RMSE scores, their mean, and standard deviation (calculated from cross-validation results)
print_scores(rmse_model_scores)
```
